```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.pipeline import Pipeline
import pickle
```

```python
df = pd.read_csv("C:/Users/chand/Downloads/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df.info()
```

```python
df['MultipleLines'] = df['MultipleLines'].str.replace('No phone service', 'No')
df['OnlineSecurity'] = df['OnlineSecurity'].str.replace('No internet service', 'No')
df['OnlineBackup'] = df['OnlineBackup'].str.replace('No internet service', 'No')
df['DeviceProtection'] = df['DeviceProtection'].str.replace('No internet service', 'No')
df['StreamingTV'] = df['StreamingTV'].str.replace('No internet service', 'No')
df['StreamingMovies'] = df['StreamingMovies'].str.replace('No internet service', 'No')
df['TechSupport'] = df['TechSupport'].str.replace('No internet service', 'No')
```

```python
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})
df['Partner'] = df['Partner'].map({'Yes': 1, 'No': 0})
df['Dependents'] = df['Dependents'].map({'Yes': 1, 'No': 0})
df['PhoneService'] = df['PhoneService'].map({'Yes': 1, 'No': 0})
df['MultipleLines'] = df['MultipleLines'].map({'Yes': 1, 'No': 0})
df['OnlineSecurity'] = df['OnlineSecurity'].map({'Yes': 1, 'No': 0})
df['OnlineBackup'] = df['OnlineBackup'].map({'Yes': 1, 'No': 0})
df['DeviceProtection'] = df['DeviceProtection'].map({'Yes': 1, 'No': 0})
df['StreamingTV'] = df['StreamingTV'].map({'Yes': 1, 'No': 0})
df['StreamingMovies'] = df['StreamingMovies'].map({'Yes': 1, 'No': 0})
df['PaperlessBilling'] = df['PaperlessBilling'].map({'Yes': 1, 'No': 0})
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
df['TechSupport'] = df['TechSupport'].map({'Yes': 1, 'No': 0})
```

```python
new_df_1 = pd.get_dummies(df['InternetService'])
new_df_2 = pd.get_dummies(df['Contract'])
new_df_3 = pd.get_dummies(df['PaymentMethod'])
df = df.join(new_df_1)
df = df.join(new_df_2)
df = df.join(new_df_3)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
```

```python
df = df.drop(columns = ['Contract', 'InternetService', 'PaymentMethod', 'customerID'])
```

```python
df.info()
```

```python
X = df.drop('Churn', axis = 1)
y = df['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.2,random_state =123)
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```

```python
X = df.drop('Churn', axis = 1)
y = df['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.2,random_state =123)
classifier = LogisticRegression(random_state = 42)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
pred = classifier.predict_proba(X)[:,1]*100
```

```python
df_pred = pd.read_csv("C:/Users/chand/Downloads/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df_pred["Predictions"] = pred
df_pred['Age'] = pred
```

```python
df_pred['StreamingTV'] = df_pred['StreamingTV'].str.replace('No internet service', 'No')
df_pred['StreamingMovies'] = df_pred['StreamingMovies'].str.replace('No internet service', 'No')
```

```python
for i in range(len(df_pred['customerID'])):
    if df_pred['SeniorCitizen'][i] == 1:
        df_pred['Age'][i] = '60 above'
    else:
        if df_pred['PaymentMethod'][i] != 'Credit card (automatic)':
            if df_pred['Partner'][i] == 'No':
                df_pred['Age'][i] = '18 below'
            elif df_pred['Partner'][i] == 'Yes':
                if df_pred['StreamingMovies'][i] == 'No' and df_pred['StreamingTV'][i] == 'Yes':
                    df_pred['Age'][i] = '40-60'
                elif df_pred['StreamingMovies'][i] == 'Yes' and df_pred['StreamingTV'][i] == 'Yes':
                    df_pred['Age'][i] = '30-40'
                elif df_pred['StreamingMovies'][i] == 'No' and df_pred['StreamingTV'][i] == 'Yes':
                    df_pred['Age'][i] = '30-40'
                elif df_pred['StreamingMovies'][i] == 'No' and df_pred['StreamingTV'][i] == 'No':
                    df_pred['Age'][i] = '40-60'
        else:
            if df_pred['Partner'][i] == 'No':
                df_pred['Age'][i] = '18-30'
            elif df_pred['Partner'][i] == 'Yes':
                if df_pred['StreamingMovies'][i] == 'Yes' and df_pred['StreamingTV'][i] == 'No':
                    df_pred['Age'][i] = '18-30'
                elif df_pred['StreamingMovies'][i] == 'Yes' and df_pred['StreamingTV'][i] == 'Yes':
                    df_pred['Age'][i] = '30-40'
                elif df_pred['StreamingMovies'][i] == 'No' and df_pred['StreamingTV'][i] == 'Yes':
                    df_pred['Age'][i] = '40-60'
                elif df_pred['StreamingMovies'][i] == 'No' and df_pred['StreamingTV'][i] == 'No':
                    df_pred['Age'][i] = '40-60'
```

```python
total_18 = 0
total_18_c = 0
total_30 = 0
total_30_c = 0
total_40 = 0
total_40_c = 0
total_60 = 0
total_60_c = 0
total_60p = 0
total_60p_c = 0
    
for i in range(len(df_pred['Age'])):
    if df_pred['Age'][i] == '18 below':
        total_18 += 1
        if df_pred['Churn'][i] == 'Yes':
            total_18_c += 1
    if df_pred['Age'][i] == '18-30':
        total_30 += 1
        if df_pred['Churn'][i] == 'Yes':
            total_30_c += 1
    if df_pred['Age'][i] == '30-40':
        total_40 += 1
        if df_pred['Churn'][i] == 'Yes':
            total_40_c += 1
    if df_pred['Age'][i] == '40-60':
        total_60 += 1
        if df_pred['Churn'][i] == 'Yes':
            total_60_c += 1
    if df_pred['Age'][i] == '60 above':
        total_60p += 1
        if df_pred['Churn'][i] == 'Yes':
            total_60p_c += 1

churn_18 = (total_18_c/total_18)*100
churn_30 = (total_30_c/total_30)*100
churn_40 = (total_40_c/total_40)*100
churn_60 = (total_60_c/total_60)*100
churn_60p = (total_60p_c/total_60p)*100
```

```python
df_churn = pd.DataFrame()
df_churn['Age group'] = ['Below 18', '18-30', '30-40', '40-60', '60 Above']
df_churn['Churn rate'] = [churn_18, churn_30, churn_40, churn_60, churn_60p]
```
